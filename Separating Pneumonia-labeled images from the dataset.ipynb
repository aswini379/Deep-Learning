# Imported the "train.csv" into a df

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
df=pd.read_csv("CheXpert-v1.0-small/train.csv")
df.head()
```

# Replaced the NaN values with '-'

```python
df=df.fillna('-')
df.head()
```

# Obtained the count by "Edema" and created a separate folder for images containing only edema labels (77,408 images)

```python
df.drop(df[df.Edema == '-'].index, inplace=True)
df['Edema'].value_counts()
```

```python
df = df[['Path','Sex','Age','Frontal/Lateral','AP/PA','Edema']]
df = df[(df['Frontal/Lateral']=='Frontal')]
df['Path'].head()
df.to_csv("train-edemaonly.csv")
```

```python
from shutil import copyfile
for p in df['Path']:
    sourcePath = p
    str = p.split('/')
    destPath = 'train/' + str[2] + '-' + str[3] + '-' + str[4]
    copyfile(sourcePath, destPath)
```
